```python
#Step 1:- First we will import the libraries we need, we will use NumPy, pandas, Matplotlib and Sklearn.
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from matplotlib import pyplot as plt
from sklearn.metrics import accuracy_score
```

```python
#Step 2:- We will be loading the iris dataset to the data frame and observe its features.
iris=load_iris()
dataframe=pd.DataFrame(iris.data,columns=iris.feature_names)
dataframe.head()
```

```python
#Step 3:- We will be preparing the train data and the test data for that we will be using the train_test_split built-in function of the sklearn.
#This function will split the data features and data-target into the train data and test data.
#For out purpose we split the dataset into 120 training and 30 testing data.
x=iris.data
y=iris.target
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
print("x train data", x_train.shape)
print("x test data", x_test.shape)
print("y train data", y_train.shape)
print("y test data", y_test.shape)
```

```python
#Step 4:- Now we create the ML model for this we wil be using Decision Tree Classifier. Which will fit the train data and test data to this classifier.
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
print("The Model is Succefully Trained")
```

```python
#Step 5:- Visualise The ML model. We will use plot_tree function to visualise the decision made by decision tree.
plt.figure(figsize=(20,10))
tree.plot_tree(model, filled=True)
```

```python
#Step 6:- Testing the model. We will use test data to get predicted value.
prediction=model.predict(x_test)
prediction 
df=pd.DataFrame({'predicted value':prediction,'actual value':y_test})
df['label']=df['predicted value'].replace(dict(enumerate(iris.target_names)))
df.head()
```

```python
#Step 7:- Checking the accuracy. We will check the accuracy by using the accuracy_score function.
score=accuracy_score(prediction,y_test)
print(f"Prediction accuracy is {score*100}%")
```

```python
#In conclusion we get 96.66666667% accuracy after testing the data. Which means our decision tree classifier works well.
```
